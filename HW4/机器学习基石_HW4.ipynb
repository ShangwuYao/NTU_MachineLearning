```python
import numpy as np

#question 13

def sign(x):
    if x > 0: return 1
    else : return -1
    
def read_data(input_file):
    f = open(input_file)
    X = []
    Y = []
    for line in f:
        strs = line.split()
        x = [1] + [float(v) for v in strs[:-1]]
        y = int(strs[-1])
        X.append(x)
        Y.append(y)
    return X, Y

def get_W(X, Y, Lambda):
    W = np.linalg.inv(np.transpose(np.array(X)).dot(np.array(X)) + Lambda * np.eye(len(X[0]))).dot(np.transpose(np.array(X))).dot(np.array(Y))
    return W
    
def test(X, Y, W):
    count = 0
    for i in range(len(Y)):
        if sign(np.array(W).dot(np.transpose(np.array(X[i])))) != Y[i]:
            count += 1
    average_error = count / len(Y)
    return average_error
    
TRAIN_DATA = 'train_data.dat'
TEST_DATA = 'test_data.dat'
# question 13
# LAMBDA = 10
LAMBDA = [pow(10, 2), pow(10,1), pow(10, 0), pow(10, -1), pow(10, -2), pow(10, -3), pow(10, -4), pow(10, -5), pow(10, -6), pow(10, -7), pow(10, -8), pow(10, -9), pow(10, -10)]
FOLDS = 5    

def generate_TrainAndVal(X, i, folds):
    n = int(len(X) / folds)
    X = np.array(X)
    train_X = np.concatenate((X[:(i * n),], X[((i + 1) * n):,]), 0)
    val_X = np.array(X)[(i * n) : ((i + 1) * n),]
    return train_X, val_X
    
def main():
    X, Y = read_data(TRAIN_DATA)
    min_Error = 1
    optimum_W = [0 for i in X[0]]
    power = 0
    # test_X, test_Y = read_data(TEST_DATA)
    for i in range(len(LAMBDA)):
        # W = get_W(train_X, train_Y, LAMBDA[i])
        
        # question 14
        #Ein = test(train_X, train_Y, W)
        #if Ein < min_Ein:
        #    min_Ein = Ein
        #    optimum_W = W
        #    power = i
        # question 15
        #Eout = test(test_X, test_Y, W)
        #if Eout < min_Eout:
        #    min_Eout = Eout
        #    optimum_W = W
        #    power = i
        # question 16
        #Etrain = test(train_X, train_Y, W)
        #if Etrain < min_Etrain:
        #    min_Etrain =Etrain
        #    optimum_W = W
        #    power = i
        # question 17
        #Eval = test(val_X, val_Y, W)
        #if Eval < min_Error:
        #    min_Error = Eval
        #    optimum_W = W
        #    power = i
        # question 19~20
        E_sum = 0
        for j in range(FOLDS):
            train_X, val_X = generate_TrainAndVal(X, j, FOLDS)
            train_Y, val_Y = generate_TrainAndVal(Y, j, FOLDS)
            W = get_W(train_X, train_Y, LAMBDA[i])
            E_sum += test(val_X, val_Y, W)
        Evc = E_sum / FOLDS
        if Evc < min_Error:
            min_Error = Evc
            power = i
        
    # question 15
    test_X, test_Y = read_data(TEST_DATA)
    # question 18
    # recalculate Weight with lambda known
    new_W = get_W(X, Y, LAMBDA[power])
    Eout = test(test_X, test_Y, new_W)
    Ein = test(X, Y, new_W)
    #Etrain = test(train_X, train_Y, optimum_W)
    # question 14
    # Eout = test(test_X, test_Y, optimum_W)
    print(LAMBDA[power])
    print(min_Error)
    print(Ein)
    print(Eout)
    
if __name__ == '__main__':
    main()
```
