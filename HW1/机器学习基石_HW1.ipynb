```python
import random
from numpy import array, inner, zeros

# Homework 18~20

def sign(x):
    if x > 0:
        return 1
    return -1

TRAIN_DATA = '18_train_data.dat'
TEST_DATA = '18_test_data.dat'

def read_data(INPUT_DATA):
    X = []
    Y = []
    f = open(INPUT_DATA)
    for lines in f:
        strs = lines.split()
        x = [1] + [float(v) for v in strs[:-1]]
        X.append(tuple(x))
        Y.append(int(strs[-1]))
    return array(X), array(Y)

def test(W, X, Y):
    n = len(Y)
    ne = sum([1 for i in range(n) if sign(inner(X[i], W)) != Y[i]])
    error = ne / float(n)
    return error
    
def train(X, Y, update_times = 50, pocket = True):
    n = len(Y)
    d = len(X[0])
    W = zeros(d)
    best_W = zeros(d)
    mistakes_best = test(best_W, X, Y)
    
    for i in range(update_times):
        random_index = random.sample(range(n), n)
        for idx in random_index :
            if sign(inner(X[idx], W)) != Y[idx]:
                W = W + Y[idx] * X[idx]
                mistakes_now = test(W, X, Y)
                if mistakes_best > mistakes_now:
                    mistakes_best = mistakes_now
                    best_W = W
                break
    
    if pocket:
        return best_W
    else: 
        return W

def pocket(test_times):
    X,Y = read_data(TRAIN_DATA)
    TX, TY = read_data(TEST_DATA)
    
    sum = 0
    for i in range(test_times):
        # question 18 
        # W = train(X, Y, pocket = True)
        # question 19  0.3899999999999887
        # W = train(X, Y, pocket = False)
        # question 20 
        W = train(X, Y, pocket = True, update_times = 100)
        error = test(W, X, Y)
        sum += error
        # print(str(i))
    aver = sum / test_times
    return aver
    
def main():
    # question 18
    aver = pocket(2000)
    # print(aver)
    
if __name__ == '__main__':
    main()
```
